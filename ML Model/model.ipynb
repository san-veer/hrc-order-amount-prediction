### 1. Checking for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

#### Boxplot before replacing outliers 

```python
plt.subplots(figsize=(10, 6))
sns.boxplot(df['amount_in_usd'])
plt.title('Before replacing outliers', fontsize=15)
plt.ylabel('Amount in USD', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().set_facecolor('#F0F0F0')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# implementing IQR

Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_indices = (df['amount_in_usd'] < lower_bound) | (df['amount_in_usd'] > upper_bound)

#replacing the outlier data with median 
df.loc[outlier_indices, 'amount_in_usd'] = df['amount_in_usd'].median()
```

### Note: Introducing a particular module for datetime manipulation

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)

    melt = melt.reset_index(drop=True)

    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff'] = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
```

```python
df = difference_in_days(df, lags=3, ffday=1, customer_id_col='Unique_Cust_ID', create_date_col='ORDER_CREATION_DATE', net_amount_col='amount_in_usd')

# checking the new columns if added
df.head()
```

#### Randomized Search of LGBM

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

regressor = lgb.LGBMRegressor()

# parameter distribution
param_dist = {
    'learning_rate': uniform(0.01, 0.3),
    'max_depth': randint(3, 13),
    'subsample': uniform(0.6, 0.4),
    'num_leaves': [2**d for d in range(3, 13)],
    'n_estimators': randint(100, 1000)
}

# performing randomized search cross-validation with hyperparameter tuning
random_search = RandomizedSearchCV(estimator=regressor, param_distributions=param_dist, scoring='neg_mean_squared_error', cv=5, n_iter=10)
random_search.fit(X_train, y_train)

# best hyperparameters and the best model
best_params = random_search.best_params_
best_model = random_search.best_estimator_

# fitting the best model on the training data
best_model.fit(X_train, y_train)

# making predictions on the training and testing data
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)
```
